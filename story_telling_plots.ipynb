```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
student_df = pd.read_csv('student_lifestyle_dataset.csv')
```

```python
student_df.head()
```

```python
# Scatter plot: GPA vs. Study Hours
plt.figure(figsize=(8, 6))
sns.scatterplot(x=student_df['Study_Hours_Per_Day'], y=student_df['GPA'], alpha=0.7)
plt.title('Correlation Between GPA and Study Hours Per Day', fontsize=14)
plt.xlabel('Study Hours Per Day', fontsize=12)
plt.ylabel('GPA', fontsize=12)
plt.grid(alpha=0.3)
plt.show()
```

```python
# Calculate the correlation between Physical Activity and Sleep Hours
correlation = student_df['Study_Hours_Per_Day'].corr(student_df['GPA'])

# Print the result
print(f"Correlation between Study_Hours_Per_Day and GPA: {correlation:.2f}")
```

***There appears to be a slight upward trend (correlation of 0.73), suggesting a potential positive correlation: students who study more tend to have higher GPAs
However, there might be diminishing returns for higher study hours (e.g., beyond ~7 hours***

```python
# Scatter plot: GPA vs. Stress Level
plt.figure(figsize=(8, 6))
sns.boxplot(x=student_df['Stress_Level'], y=student_df['GPA'])
plt.title('GPA Across Different Stress Levels', fontsize=14)
plt.xlabel('Stress Level', fontsize=12)
plt.ylabel('GPA', fontsize=12)
plt.grid(alpha=0.3)
plt.show()
```

***Students with Low Stress seem to have slightly higher GPAs on average compared to those with Moderate or High Stress levels.***

```python
#Scatter plot: Extracurricular Hours vs. GPA
plt.figure(figsize=(8, 6))
sns.scatterplot(x=student_df['Extracurricular_Hours_Per_Day'], y=student_df['GPA'], alpha=0.7)
sns.regplot(x=student_df['Extracurricular_Hours_Per_Day'], y=student_df['GPA'], scatter=False, color='red', ci=None)
plt.title('Relationship Between Extracurricular Hours and GPA', fontsize=14)
plt.xlabel('Extracurricular Hours Per Day', fontsize=12)
plt.ylabel('GPA', fontsize=12)
plt.grid(alpha=0.3)
plt.show()
```
